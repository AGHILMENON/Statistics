```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew,kurtosis
```

```python
cars=pd.read_csv('Q9_b.csv') #read Q9.csv file)
cars
```

```python
SP=cars['SP']
WT=cars['WT']
```

```python
#Initialising skew and Kurtosis value for a term
```

```python
Skewness_SP=cars['SP'].skew()
Skewness_SP
```

```python
Skewness_WT=cars['WT'].skew()
Skewness_WT
```

```python
Kurtosis_SP=cars['SP'].kurtosis()
Kurtosis_SP
```

```python
Kurtosis_WT=cars['WT'].kurtosis()
Kurtosis_WT
```

```python
# Plotting the distribution of Car SP
```

```python
plt.figure(figsize=(8,6))
plt.plot(SP,label=f"Skewness: {Skewness_SP:.2f},Kutosis: {Kurtosis_SP:.2f}")
plt.title('Car SP')
plt.xlabel('Data Point')
plt.ylabel('SP')
plt.legend()
plt.show()
```

```python
# Plotting the distribution of Car SP
```

```python
plt.figure(figsize=(8,6))
plt.plot(WT,label=f"Skewness: {Skewness_WT:.2f},Kutosis: {Kurtosis_WT:.2f}")
plt.title('Car WT')
plt.xlabel('Data Point')
plt.ylabel('WT')
plt.legend()
plt.show()
```

# Inferences:  
a)	SP: From the above plot and distribution, for speed we got Skewness=1.61 and Kurtosis=2.98…. from that since skewness value is positive, we can infer that distribution is right skewed with longer tails on right end and regarding Kurtosis it is leptokurtic …there are more extreme values or outliers in the dataset with higher tail ends compared to normal distribution.


b)	WT: From the above plot and distribution, for distance we got Skewness=
-0.61 and Kurtosis=0.95…. from that we can infer since skewness value is slightly negative, we can say distribution is slightly left skewed and longer tail on left end and regarding Kurtosis it is Leptokurtic…. there are many extreme values or outliers in the dataset with heavy tails compared to normal distribution.
