```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={'Allied SIgnal':24.23,"Banker's Trust":24.53,"General Mill's":25.41,"ITT Industries":24.14,"J.P.Morgan & Co.":29.62,"Lehman Brothers":28.25,
      "Marriott":25.81,"MCI":24.39,"Merrill Lynch":40.26,"Microsoft":32.95,"Morgan Stanley":91.36,"Sun Microsystems":25.99,"Travelers":39.42,
      "US Airways":26.71,"Warner-Lambert":35.00}
data
```

```python
keys=data.keys()
```

```python
Company_name=list(keys)
Company_name
```

```python
values=data.values()
```

```python
Measure=list(values)
Measure
```

```python
plt.title("COMPANY VS MEASURES")
plt.plot(Company_name,Measure,marker='D')
plt.xlabel('Company Name')
plt.xticks(rotation=60)
plt.ylabel('Measures')
plt.show()
```

Mean

```python
mean_measure=sum(Measure)/len(Measure)
mean_measure
```

Variance

```python
sum_Numerator=0
for i in Measure:
    Numerator=(i-mean_measure)**2
    sum_Numerator +=Numerator
Variance=sum_Numerator/(len(Measure))
Variance
```

Standard Deviation

```python
Standard_deviation=np.sqrt(Variance)
Standard_deviation
```
